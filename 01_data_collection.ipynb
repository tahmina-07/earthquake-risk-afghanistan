```python
import requests
import pandas as pd
import os 

print("Libraries imported successfully")
```

```python
# Afg coordinates 

MIN_LAT = 29.0 # Southern Border 
MAX_LAT = 39.0 # Northern Border 
MIN_LON = 60.0 # Western Border 
MAX_LON = 75.0 # Eastern Border

# Earthquake from 1990-present 
START_DATE = "1990-01-01"
END_DATE = "2024-12-31"

# Only earthquakes magnitudes 4.0 and above 
MIN_MAGNITUDE = 4.0

print(f"Searching: Magnitude {MIN_MAGNITUDE}+ earthquakes")
print(f"Region: Afghanistan bounding box")
print(f"Period: {START_DATE} to {END_DATE}")
```

```python
# URL that asks USGS for earthquake date for our coordinates 

BASE_URL = "https://earthquake.usgs.gov/fdsnws/event/1/query"
params = {
    "format": "csv",          # get data as a CSV table
    "starttime": START_DATE,
    "endtime": END_DATE,
    "minlatitude": MIN_LAT,
    "maxlatitude": MAX_LAT,
    "minlongitude": MIN_LON,
    "maxlongitude": MAX_LON,
    "minmagnitude": MIN_MAGNITUDE,
    "orderby": "time"  
}

print("API parameters ready!")
print(f"URL will be: {BASE_URL}")
```

```python
print("Contacting USGS API... this may take 30-60 seconds")

response = requests.get(BASE_URL, params=params)

if response.status_code == 200:
    print(f"✅ Success! Data received from USGS")
    print(f"Data size: {len(response.content) / 1024:.1f} KB")
else:
    print(f"❌ Error: {response.status_code}")
    print(response.text)
```

```python
import io 

df = pd.read_csv(io.StringIO(response.text))

os.makedirs("../data", exist_ok=True)
df.to_csv("../data/usgs_afghanistan_raw.csv", index=False)

print(f"✅ Dataset saved!")
print(f"Total earthquakes found: {len(df)}")
print(f"\nColumns in dataset:")
print(df.columns.tolist())
print(f"\nFirst 3 rows:")
df.head(3)
```

```python
print("---Dataset Overview---")
print(f"Total earthquakes: {len(df)}")
print(f"Date range: {df['time'].min()} to {df['time'].max()}")
print(f"Magnitude range: {df['mag'].min()} to {df['mag'].max()}")
print(f"\nTop 5 largest earthquakes:")
df[['time', 'latitude', 'longitude', 'mag', 'depth', 'place']].sort_values('mag', ascending=False).head()
```
